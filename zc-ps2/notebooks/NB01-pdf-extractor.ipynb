# NB01 - Extract Text and Chunk PDF files

```python
import sys
import os
import glob
import json
from tqdm.notebook import tqdm

# Add the project root to Python path (assuming current working directory is 'notebooks')
project_root = os.path.abspath(os.path.join(os.getcwd(), '..'))
if project_root not in sys.path:
    sys.path.insert(0, project_root)

from utils import extract_text_data_from_pdf
```

## 1. Identify all PDF files

```python
# Glob all files under ../data/pdfs/
files = glob.glob(os.path.join("..", "data", "pdfs", "*"))
len(files)
```

## 2. Parse PDFs


```python
# TODO: You might want to do this in batches if you have too many files, or it will take a long time.
doc_text = {os.path.splitext(os.path.basename(file))[0]: extract_text_data_from_pdf(file, unstructured_strategy="fast")
            for file in tqdm(files)}

len(doc_text)
```

Note that the unstructured `partition_pdf()` function already collects a lot of metadata about each chunk of PDF. These will be used in later steps.
